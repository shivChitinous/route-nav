# The Goal

We can generate random scenes in the world. Using these random views and a Foldiak-network model, we learn a sparse set of bases in which to view the world. We can then compare naive responses of KCs to learnt responses and ask if the sparse representations do a better job of seperating them out.

```python
n = 100 #number of snaps
```

```python
snapLocs = snp.snapLocs(dWorld, zAnt, n)
```

```python
visibility = 2 #in m
hfovDeg = 296 #deg
resDeg = 4 #deg
upDeg = 60 #deg
downDeg = -15 #deg
```

Generate a scene

```python
frame = 50
```

```python
scene = pr.generateScene(0, frame, visibility, dWorld, snapLocs, 
                         hfovDeg, resDeg, upDeg, downDeg, zAnt)
```

```python
plt.imshow(scene,cmap='gray',origin='lower');
```

## Collect all Scenes

```python
scenes = pr.walkThrough(n, 0, n,
               visibility, dWorld, snapLocs, hfovDeg, resDeg, upDeg, downDeg, zAnt, animate=True)
```
